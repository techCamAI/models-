This File was used in order to convert the Medical Mask Dataset to required YOLO format after which augmentation took place in
roboflow.... in the process some filtering also took place in order to have representation of onlY 2 classes

```python
import os, cv2 as cv, shutil
import json as Json

class conversion(object):
    def __init__(self, *args, **kwargs):
        pass

    def ImageBbox(self, JS:str, image:str):
        # basic Json reading structure
        data = open(JS)
        reading = Json.load(data)

        # extracting required info
        allowed_classes = ['face_no_mask', 'face_with_mask_incorrect'] # used for later purposes
        for each in reading['Annotations']:
            if  ('face_no_mask' or 'face_with_mask_incorrect') in [lst['classname'] for lst in reading['Annotations']]:
                    #or 'face_with_mask_incorrect' in [lst['classname'] for lst in reading['Annotations']]:
                print('ess')
                data = each['BoundingBox']

                # reading data and converting to requirede
                lX, tY, rX, bY = data
                w, h = rX - lX, bY - tY
                cX, cY = lX + (w/2), tY + (h/2)

                # image dimensions
                H, W, _ = cv.imread(image).shape

                # final normalization for YOLO format

                # center coords
                cX /= W
                cY /= H

                # width and height
                w /= W
                h /= H
                lst = ['hijab_niqab', 'mask_colorful', 'mask_surgical', 'face_no_mask', 'face_with_mask_incorrect', 'face_with_mask', 'face_other_covering', 'scarf_bandana', 'balaclava_ski_mask', 'face_shield', 'other', 'gas_mask', 'turban', 'helmet', 'sunglasses', 'eyeglasses', 'hair_net', 'hat', 'goggles', 'hood', ]
                classN = lst.index(each['classname'])

                string_to_append = f"{classN} {cX} {cY} {w} {h}"
                image_file = image[-8:-4]


                with open(f'Mask Dataset/annotations/{image_file}.txt', 'a') as file:
                    file.write(f'{string_to_append}\n')
                    file.close()

                # pasting image
                print(image)
                shutil.copy(image, r'Mask Dataset/Images')
```

```python
"""the installed dir from which the images were first takes is deleted from the root dir
so running this will throw an error"""
images, Js = os.listdir('Medical mask/Medical Mask/images'), os.listdir('Medical mask/Medical Mask/annotations')
initalized_converter = conversion()
which_files = [f"{'0' * (4 - len(str(num)))}{num}" for num in range(1, 6400)]
```

```python
for index, each in enumerate(which_files):
    try:
        initalized_converter.ImageBbox(f'Medical mask/Medical Mask/annotations/{each}.json', f'Medical mask/Medical Mask/images/{each}.jpg')
        print('done', index)
    except FileNotFoundError as e:
        print(e)
```
